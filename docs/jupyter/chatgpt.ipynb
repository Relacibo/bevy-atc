```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# Gegebene Werte
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([0, 0.05, 0.23, 0.5, 0.9, 1.4, 2])

# Quadratische Regression (Polynom 2. Ordnung)
coeffs = Polynomial.fit(x, y, 2).convert().coef
print("Gefittete Koeffizienten (a, b, c) f√ºr y = a + b*x + c*x^2:")
print(coeffs)

# Werte zum Zeichnen der Kurve
x_fit = np.linspace(0, 6, 100)
y_fit = coeffs[0] + coeffs[1]*x_fit + coeffs[2]*x_fit**2

# Plot
plt.scatter(x, y, color='red', label='Originaldaten')
plt.plot(x_fit, y_fit, color='blue', label='Quadratische Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratische Anpassung')
plt.legend()
plt.grid(True)
plt.show()
```
